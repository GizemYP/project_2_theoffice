```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.imdb.com/list/ls000977475/?sort=release_date,asc&st_dt=&mode=detail&page=1&ref_=ttls_vm_dtl/'
url2 = 'https://www.imdb.com/list/ls000977475/?st_dt=&mode=detail&page=2&ref_=ttls_vm_dtl%2F&sort=release_date,asc'
```

```python
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
articles = soup.find_all('span', class_='ipl-rating-star__rating')
titles = soup.select(".lister-item-header a:nth-of-type(2)")
syn = soup.find_all('div',class_="lister-item-content")
```

```python
imdb_rating = []
imdb_title = []
synopsis = []
```

```python
x = 0
while x<len(articles):
    imdb_rating.append(articles[x].text)
    x+=23
```

```python
for t in titles:
    imdb_title.append(t.text)
```

```python
for s in syn:
    syn_ep = s.find('p',class_="")
    synopsis.append(syn_ep.text.replace("\n    ",""))
```
